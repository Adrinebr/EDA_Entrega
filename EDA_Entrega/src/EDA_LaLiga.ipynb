```python
plt.figure(figsize=(20, 10))
sns.countplot(x="Campeón", hue="Campeón",data=df_LaLiga, order=df_LaLiga["Campeón"].value_counts().index)
plt.title("Campeones de Liga")
plt.xlabel("Equipo")
plt.ylabel("Número títulos de Liga")
plt.legend(title="Campeón")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
ax = sns.countplot(x="Campeón", hue="Campeón", data=df_LaLiga, order=df_LaLiga["Campeón"].value_counts().index)
plt.title("Campeones de Liga")
plt.xlabel("Equipo")
plt.ylabel("Número títulos de Liga")

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
campeon_y_pichichi_df = filas_con_real_madrid[(filas_con_real_madrid['Campeón'] == 'Real Madrid C. F.') & (filas_con_real_madrid['Equipo Pichichi'] == 'Real Madrid C. F.')]
campeon_y_pichichi_df
```
